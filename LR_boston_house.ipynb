```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
boston_df = pd.read_csv('/content/drive/MyDrive/ML Projects/BOSTON/Boston.csv')
```

```python
boston_df
```

```python
#prepare dataset
```

```python
dataset = boston_df.drop(['Unnamed: 0','medv'], axis= 1)
dataset
```

```python
dataset['Price'] = boston_df['medv']
dataset
```

```python
dataset.info()
```

```python
## independent and dependent feature
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
# train test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
## standardizing dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
# Model Training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
#### print coefficients and intercept

print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
### On which parameter model has been trained

regression.get_params()
```

```python
### prediction with Test data

test_pred = regression.predict(X_test)
test_pred
```

```python
## plot a scatter plot for the prediction
```

```python
plt.scatter(y_test, test_pred)
```

```python
residual = y_test - test_pred
```

```python
## plotting this residual

sns.displot(residual)
```

```python
sns.displot(residual, kind = 'kde')
```

```python
# scatter plot wrt to prediction and residuals

plt.scatter(test_pred, residual)

# kind of uniform distribution
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print("Mean Squared Error:",mean_squared_error(y_test,test_pred))
print("Root Mean Squred Error:", np.sqrt(mean_squared_error(y_test,test_pred)))
print("Mean Absolute Error:", mean_absolute_error(y_test, test_pred))
```

```python
# R squared and Adjusted R squre metric
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, test_pred)
score
```

```python
#Adjusted R2 = 1 â€“ [(1-R2)*(n-1)/(n-k-1)]

#where:

#R2: The R2 of the model n: The number of observations k: The number of predictor variables
```

```python
from sklearn.metrics.cluster import adjusted_rand_score
adjusted_r2_score = 1 - ((1-score)*(len(y_test) - 1)/ (len(y_test) - X_test.shape[1] - 1))
adjusted_r2_score
```
